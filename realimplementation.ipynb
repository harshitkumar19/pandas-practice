```python
#total runs scored by each team across all seasons
df1.groupby('batting_team')['total_runs'].sum()
```

```python
#average runs scored per innings by each team
df1.groupby('batting_team')['total_runs'].mean()
```

```python
#win percentage of each team
matches_played = df1[['match_id', 'batting_team']].drop_duplicates()
matches_played_count = matches_played['batting_team'].value_counts()
matches_won = df1[['match_id', 'winner']].drop_duplicates()
matches_won_count = matches_won['winner'].value_counts()
win_percentage = (matches_won_count / matches_played_count) * 100
win_percentage
```

```python
#Analyze toss impact:
#how many times toss winner also won the match
temp = df1[['match_id','toss_winner','winner']].drop_duplicates()
temp['toss_win_match_win'] = temp['toss_winner'] == temp['winner']

temp['toss_win_match_win'].value_counts(normalize=True) * 100

```
