# MISSING DATA
---

````python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
````

````python
df
````

````python
df.isna()
````

````python
df.isna().sum()
````

````python
# The any() method is used to check if there are any True values in the DataFrame.
# In this case, it will return True for each column that contains at least one NaN value, and False for columns that do not contain any NaN values.
df.isna().any()
````

````python
# dropna() is a method in pandas that is used to remove missing values (NaN) from a DataFrame. 
# By default, it removes any row that contains at least one NaN value. You can also specify to drop columns 
# instead of rows by using the axis parameter.
df.dropna()
````

````python
df
````

````python
#thresh parameter is used to specify the minimum number of non-NA values required to keep a row or column.
# In this case, thresh=3 means that a row will be dropped if it has less than 3 non-NA values.
df.dropna(thresh=3)
````

### Filling the missing data with fillna() method
---

````python
#fillna(value) is a method in pandas that is used to fill missing values (NaN) in a DataFrame with a specified value. 
# This can be useful for handling missing data and ensuring that your DataFrame does not contain any NaN values.
df.fillna(0)
````

````python
values= {'A' :0,'B':10,'C':50,'D':200}
df.fillna(value=values)
````

````python
df
````

````python
df.fillna(df.mean())
````

## MERGE, JOIN AND CONCATENATE
--- 

````python

employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
````

````python
employees
````

````python
salaries
````

````python
pd.merge(employees, salaries)
````

````python
#inner join: This is the default join type. It returns only the rows that have matching values in both DataFrames based on the specified key column(s). 
# In this case, it will return only the employees who have a corresponding salary record in the salaries DataFrame.
pd.merge(employees,salaries,on='employee_id',how ='inner')
````

````python
# outer join: This join type returns all rows from both DataFrames,
# and fills in NaN for missing matches on either side.
pd.merge(employees,salaries,on='employee_id',how ='outer')
````

````python
# left join: This join type returns all rows from the left DataFrame (employees) and the matching rows from the right DataFrame (salaries). 
# If there is no match, the result will contain NaN for the columns from the right DataFrame.
pd.merge(employees,salaries,on='employee_id',how ='left')
````

````python
#right join: This join type returns all rows from the right DataFrame (salaries) and the matching rows from the left DataFrame (employees). 
# If there is no match, the result will contain NaN for the columns from the left DataFrame.
pd.merge(employees,salaries,on='employee_id',how ='right')
````

**grouping**

````python

data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
````

````python
df
````

````python
#grouping is a powerful operation in pandas that allows you to split your data into groups based on one or more columns, 
# and then apply a function to each group independently. This is often used for aggregation, transformation
cat = df.groupby('Category')
cat
for i,v in cat:
    print(i)
    print(v)
````

````python
cat = df.groupby('Category')['Sales'].sum()
cat
````

````python
cat = df.groupby('Store')['Sales'].sum()
cat
````

````python
# group by multiple columns
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
````

**PIVOT TABLE**

````python

data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
````

````python
pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='sum')
````

````python
pi =pd.pivot_table(df,values=['Sales','Units'],index='Region',columns='Product',aggfunc='sum')
pi
````

**Cross Tabulation**
---
the basic diffrence between pivot table and cross tabulation is that pivot table can take multiple aggregation function but cross tabulation can take only one aggregation function.

````python
newdf =pd.crosstab(df['Region'],df['Product'])
newdf
````

````python
newdf =pd.crosstab(df['Region'],df['Product'],values=df['Sales'],aggfunc='sum')
newdf
````
