```python
# Dataframe uses in 2d data
data =  { 'Name': ['Harshit','alien','kansika'],
    'City':['Berlin','Null','Naubasta'],
    'Country': ['Germany','Greenland','America'],
    'Age': ['24','35','60'] }
df = pd.DataFrame(data)
df
```

```python
data =  [ ['Harshit','alien','kansika'],
           ['Berlin','Null','Naubasta'],
           ['Germany','Greenland','America'],
            ['24','35','60'] ]
df = pd.DataFrame(data,columns=['Name','Country','Age'])
df
```

```python
data =  [['Harshit','alien','kansika'],
         ['Berlin','Null','Naubasta'],
         ['Germany','Greenland','America'],
         ['24','35','60'] ]
df = pd.DataFrame(data,
                    columns=['Name','Country','Age'],
                    index=['person1','person2','person3','person4']
                    )

df
```

```python
# Dataframe uses in 2d data
data =  {
    'Joe':[10,24,35,],
     'harry':[20.00],
     'Patlu':[24.4]   
}

df = pd.DataFrame(data.items(),columns =['Name','Marks'])
df
```

```python
data =  {
    'Joe':[10,24,35],
     'harry':[20.00,45,50],
     'Patlu':[24.4,0.00,60]      
}
print(data )

t = pd.DataFrame(data).T
print(t)
```

```python
students_info = {
    'Name': ['Harshit','alien','kansika'],
    'City':['Shahjahanpur','kanpur','Naubasta'],
    'Country': ['Kanpur','kannauj','america'],
    'Age': ['24','35','60']
}
df = pd.DataFrame.from_dict(
    students_info, 
    orient='index',
    columns=['Student1','Student2','Student3'])

df
```
